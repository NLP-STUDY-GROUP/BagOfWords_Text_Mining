## 6.1 차원의 저주와 차원 축소의 이유
* 차원의 저주 
    * 낮은 차원에서 발생하지 않던 문제가 차원이 커지면서 발생하는 문제를 의미
    * 차원의 크기 커짐
        * --> 관측한 변수의 수가 증가하는 것을 의미
    * 차원이 커지면
        * --> 차원이 구성하는 공간의 크기가 커짐으로써 
        * --> 공간 안에 있는 데이터들 간의 거리가 멀어짐
        * --> 결과, 데이터가 희박해지는 문제 --> 차원의 저주 
        * 같은 공간 안에서 데이터의 수는 동일, 데이터 간의 거리가 멀어질 수 밖에 없음  
    * 데이터가 희박해지면 
        * --> 이러한 데이터를 기반으로 학습한 모형의 설명력이 떨어짐 
        * --> 통계적 유의성을 요구하는 모형의 경우, 요구되는 데이터의 크기가 기하급수적으로 커짐 
        * --> 데이터 간 거리에 기반한 방법론(ex.KNN)과 같은 알고리즘의 성능이 저하됨 
    * BOW 기반의 문서 표현 
        * 문서를 표현한느 데 사용하는 특성 --> 단어 
        * 말뭉치에 단어가 매우 많기 때문'
        * 그럼에도 불구하고 BOW기반의 기법들이 나름의 좋은 성능 유지 
    * 차원의 저주를 해결하는 방법 
        * 1. 데이터를 충분히 늘리는 것 
            * 차원이 커지며, 문제는 데이터 수가 차원의 수보다 작아지는 시점 
                * ex, 연립방정식- 미지수가 방정식의 수보다 많은 경우 
                * 해를 구하는 것이 불가능해지고, 
                * 주어진 데이터를 설명하는 모형을 찾을 수 없게됨 
                * --> 충분한 수의 문서를 확보하는 것이 중요함 
                * But, 문서 분류, 학습 데이터 (문서 수 < 문서 표현의 사용한 특성의 수) 인 경우도 잘 작동 
        * 2. BOW로 표현한 문서의 특성 수를 줄임으로써 차원을 축소하는 것 
            * 2-1. 특성 선택
                * 현재 있는 특성들 중에서 대표 특성을 선택하는 것 
                * 문서 분류 - 라쏘회귀 
                    * 계수 0이 되면, 사실상 그 특성은 결과에 영향을 미치지 못함 
                    * 아예 입력으로 사용할 필요가 없음 
                 * 단어가 나타난 문서의 수를 기반으로 너무 적거나 많은 경우에 특성에서 제외하는 방법 
            * 2-2. 특성 추출 
                * 기존의 특성 값을 조합해 새로운 특성을 생성하는 것 
                * 새로운 특성 -기존 특성들의 선형 결합으로 표현됨 
                * ex. PCA, LSA와 같은 방법들

## 6.2 PCA를 이용한 차원 축소
* PCA (Principal Component Analysis)
    * 주성분 분석 
    * 데이터의 분산을 최대한 보존하는 새로운 축을 찾아 변환 
    * --> 차원을 축소하는 방법

* PCA를 이해하기 위해 
    * 통계, 선형대수   
        * 고윳값, 고유벡터, 공분산행렬과 같은 선형대수 개념 필요 

* PCA 방법 
    * 차원을 줄이는 과정에서 정보가 최대한 유지되어야 함 
    * ex. 10개의 독립변수를 9개로 줄여도, 9개가 정보량을 유지해야 함 
    * 통계적으로 이렇나 정보량 --> 분산으로 측정 
    * 차원이 줄어들 때, 최대한 분산을 크게 유지 -->정보량의 손실을 최소화할 수 있음 
    * 분산이 커짐 --> 점들이 넓게 분포하면 
        * -->점이 겹쳐질 확률도 줄어들게 됨 
        * --> 각 점 사이의 거리도 멀어져서 점을 더 잘 분별할 수 있음 
        * --> 정보량이 가장 많다고 할 수 있음 
    * **분산을 가장 크게** 만드는 새로운 축을 찾는 방법론

```python
#-----------------------------------------------------
# PCA 실습 
    # 20 뉴스그룹 문서 가져옴 
#-----------------------------------------------------
from sklearn.datasets import fetch_20newsgroups

#20개의 토픽 중 선택하고자 하는 토픽을 리스트로 생성
categories = ['alt.atheism', 'talk.religion.misc', 'comp.graphics', 'sci.space']

#학습 데이터셋을 가져옴
newsgroups_train = fetch_20newsgroups(subset='train',
#메일 내용에서 hint가 되는 부분을 삭제 - 순수하게 내용만으로 분류
                                      remove=('headers', 'footers', 'quotes'),
                                      categories=categories)
#검증 데이터셋을 가져옴
newsgroups_test = fetch_20newsgroups(subset='test', 
                                     remove=('headers', 'footers', 'quotes'),
                                     categories=categories)
                                    
X_train = newsgroups_train.data   #학습 데이터셋 문서
y_train = newsgroups_train.target #학습 데이터셋 라벨

X_test = newsgroups_test.data     #검증 데이터셋 문서
y_test = newsgroups_test.target   #검증 데이터셋 라벨
```

```python
#-----------------------------------------------------
# PCA 실습 
    # 문서 분류를 위한 데이터 준비 
    # 토큰화, 불용어처리, 스테밍 등 전처리 이후 BOW 기반의 특성 벡터로 변환  
#-----------------------------------------------------

#--------------------------------------------불용어
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
cachedStopWords = stopwords.words("english")
english_stops = set(stopwords.words('english')) #영어 불용어를 가져옴

#--------------------------------------------토큰화 
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import RegexpTokenizer

RegTok = RegexpTokenizer("[\w']{3,}") # 정규포현식으로 토크나이저를 정의

#--------------------------------------------스테밍 
from nltk.stem.porter import PorterStemmer

#-------------------------------------------- 문서 전처리 function 생성 
def tokenizer(text):
    tokens = RegTok.tokenize(text.lower()) #이렇게 해도 되는지 확인
    # stopwords 제외
    words = [word for word in tokens if (word not in english_stops) and len(word) > 2]
    # portr stemmer 적용
    features = (list(map(lambda token: PorterStemmer().stem(token),words)))
    return features

tfidf = TfidfVectorizer(tokenizer=tokenizer)
X_train_tfidf = tfidf.fit_transform(X_train) # train set을 변환
X_test_tfidf = tfidf.transform(X_test) # test set을 변환
```

```python
#-----------------------------------------------------
# PCA 실습 
    # 차원을 축소한 뒤의 분류 성능과 비교하기 위해 
    # 사이킷런의 로지스틱 회귀분석 라이브러리 이용해 차원 축소 전 분류 성능 미리 측정 
#-----------------------------------------------------

from sklearn.linear_model import LogisticRegression 

LR_clf = LogisticRegression() #분류기 선언
LR_clf.fit(X_train_tfidf, y_train) # train data를 이용하여 분류기를 학습
print('#Train set score: {:.3f}'.format(LR_clf.score(X_train_tfidf, y_train)))
print('#Test set score: {:.3f}'.format(LR_clf.score(X_test_tfidf, y_test))) 
```

```python
#-----------------------------------------------------
# PCA 실습 
    # 사이킷런은 주성분 분석을 지원하기 위해 PCA라이브러리를 제공 
    # 매개변수 
        # n_components 
            # 축소하고자 하는 차원의 크기를 지정
        # svd_solver
            # 알고리즘을 선택할 수 있음
            # auto: 기본 값 
            # 축소 전 차원, 축소 목표 차원 고려해 자동으로 선택 
        # explained_variance
            # 객체 속성값 
            # 각각의 새로운 축이 설명한느 분산 나타냄 
        # explained_variance_ratio_ 
            # 이를 축소 전 분산에 대한 비율로 표시
            # 새로운 축들이 원래 분산을 모두 설명해주면 합은 1이됨  
    # 사이킷런 PCA
        # --> 희소벡터 형식에 대해 직접 연산을 지원하지 않음 
        # CountVectorize /TfidfVectorizer 로 변환한 행렬을 직접 인수로 사용할 수 없음
        # 따라서, toarray() 메서드 이용해 형태 변환한 후 
        # transform() 이나 fit_transform() 메서드에 인수로 전달 
        # 축소한 뒤, 원래의 분산의 어느 정도를 설명하는지 보기 위해 explained_variance_ratio_ 출력
#-----------------------------------------------------
from sklearn.decomposition import PCA

pca = PCA(n_components=2000, random_state=7) # 20,085 치원 --> 2,000차원으로 축소
X_train_pca = pca.fit_transform(X_train_tfidf.toarray())
X_test_pca = pca.transform(X_test_tfidf.toarray())

print('Original tfidf matrix shape:', X_train_tfidf.shape) 
print('PCA Converted matrix shape:', X_train_pca.shape)
print('Sum of explained variance ratio: {:.3f}'.format(pca.explained_variance_ratio_.sum()))

# 결과 해석
    # 차원이 20,085 - 특성 벡터의 단어 수를 나타냄 
    # 축소된 차원은 목표한 바와 같이 2,000
    # 원래 차원의 수가 커서 계산량이 상당히 많고, 그로 인해 시간도 꽤 걸림 
    # 시간은 걸렸으나, 차원이 1/10로 줄었음에도, 설명되는 분산 거의 100%
    # 정보의 손실이 거의 없음 
    # 축소된 데이터를 이용해 문서 분류를 함으로써 성능의 변화 체크 
```

```python
#-----------------------------------------------------
# PCA 실습 - 특성 축소  
    # 특성 축소 결과 
        # 축소된 데이터를 이용해 문서 분류를 함으로써 성능의 변화 체크 
        # 축소하기 전 성능과 동일 
        # PCA는 선형결합을 통해 최대한의 정보를 유지 
        # 특성 선택과는 다른 성능을 냄 
        # 5장 문서 분류 - 특성의 수 2,000개로 제한했을 때, test 성능 (0.734)보다 좋음 
#-----------------------------------------------------
LR_clf.fit(X_train_pca, y_train)
print('#Train set score: {:.3f}'.format(LR_clf.score(X_train_pca, y_train)))
print('#Test set score: {:.3f}'.format(LR_clf.score(X_test_pca, y_test)))
```

```python
#-----------------------------------------------------
# LASSO 실습 - 특성 선택 
    # 특성 선택 결과
        # 최종적으로 사용된 특성 321개
        # 테스트 집합에 대한 성능: 0.718
        # 특성 축소에 비해 떨어짐 
        # PCA와 비교, 목표 차원을 라쏘 회귀분석과 동일한 321개로 설정 후 비교 
#-----------------------------------------------------
lasso_clf = LogisticRegression(penalty='l1', solver='liblinear', C=1) 
lasso_clf.fit(X_train_tfidf, y_train) 

print('#Train set score: {:.3f}'.format(lasso_clf.score(X_train_tfidf, y_train)))
print('#Test set score: {:.3f}'.format(lasso_clf.score(X_test_tfidf, y_test)))

import numpy as np
# 계수(coefficient) 중에서 0이 아닌 것들의 개수를 출력
print('#Used features count: {}'.format(np.sum(lasso_clf.coef_ != 0)), 'out of', X_train_tfidf.shape[1]) 
```

```python
#-----------------------------------------------------
# LASSO 실습 vs PCA 실습 비교
    # PCA - 특성 축소 
    # LASSO - 특성 선택 
        # PCA - 설명되는 분산 43.7%로 매우 줄었지만 
        # 테스트 집합에 대한 정확도 75.1%,. 축소전 76.1에 비해 1% 줆
        # 라쏘 회귀분석 & 축소 차원 동일하지만 분류기 성능 PCA가 더 뛰어났음 
#-----------------------------------------------------
pca = PCA(n_components=321, random_state=7)

X_train_pca = pca.fit_transform(X_train_tfidf.toarray())
X_test_pca = pca.transform(X_test_tfidf.toarray())
print('PCA Converted X shape:', X_train_pca.shape)
print('Sum of explained variance ratio: {:.3f}'.format(pca.explained_variance_ratio_.sum()))

LR_clf.fit(X_train_pca, y_train)
print('#Train set score: {:.3f}'.format(LR_clf.score(X_train_pca, y_train)))
print('#Test set score: {:.3f}'.format(LR_clf.score(X_test_pca, y_test)))
```

## 6.3 LSA를 이용한 차원 축소와 의미 파악
* LSA (Latent Semantic Analysis)
    * 잠재 의미 분석
    * 문서들에 잠재된 의미를 분석하고 동시에 단어들에 잠재된 의미도 분석함 
    * 의미: 문서와 단어를 연결하는 매개체 --> 축소된 차원 
    * LSA는 SVD(Singular Value Decomposition) 
        * 절단된 SVD(Truncated SVD)를 이용해 구현됨

* SVD (Singular Value Decomposition)
    * 특잇값 분해는 m x n 의 크기를 가진 주어진 행렬을 아래와 같이 세 개의 행렬의 곱으로 분해 
    * X = U x S x (V^T)
    * U(mxm), V^T (nxn)의 크기를 갖는 직교행렬이어야 하고 
    * S 은 mxn 크기를 갖는 대각행렬이어야 한다 
    * 특잇값 분해는 위에서 설명한 PCA와 같이 결과적으로 공분산행렬의 고윳값을 만들어내고
    * 이를 통해 차원축소가 가능하게 해줌 

* 차원축소 
    * 세 개의 행렬의 곱으로 분해하는지? 
    * 분해된 세 개의 행렬을 다시 곱해 원래의 데이터로 복원할 수 있음 
    * 위 행렬에서 차원 축소의 키가 되는 행렬은 S
    * S 에는 주성분 분석 결과와 동일한 정보가 들어 있음 
    * 즉, 가장 정보량이 많은 축이 위에 있고 
    * 아래로 점차 줄어드는 형태로 새로운 축에 대한 정보가 들어감 

```python
# SVD 
    # https://angeloyeo.github.io/2019/08/01/SVD.html
    # https://ko.wikipedia.org/wiki/%ED%8A%B9%EC%9E%87%EA%B0%92_%EB%B6%84%ED%95%B4
    # https://www.fun-coding.org/recommend_basic6.html
    # https://rfriend.tistory.com/185

from IPython.display import Image
display(Image(filename= './img/ch6/svd.png', width=500, height=500 )) # help(Image)

from IPython.display import Image
display(Image(filename= './img/ch6/svd1.png',width=500, height=300 )) # help(Image)

from IPython.display import Image
display(Image(filename= './img/ch6/svd2.png',width=500, height=300)) # help(Image)
```

```python
# LSA 에서의 차원 축소 
    # SVD의 변형된 절단된 SVD(Truncated SVD)를 통해 이루어짐 
    # SBD & 절단된 SVD 의 차이 
    # 절단된 SBD
        # 차원 축소를 위해 S의 대각 원소인 특잇값에서 상위 k 개만 골라냄 
        # 결과, U에서 K개의 열 그리고 (V^T)에서의 k개의 행만 남게 됨 
        # SVD에서는 세 행렬의 곱으로 원래의 X를 복원할 수 있지만
        # 절단된 SVD에서는 완전한 복원이 불가능, 대신 최대한 유사한 값이 되도록 함 
        # 즉, 정보의 손실 발생  
        # X = Xk = UkSk(Vk ^ T)

# LSA를 사용하는 목적 
    # 1. 차원의 축소
        # 즉, X가 각 문서에 대한 카운트 벡터 / tfidf 벡터로 이루어진 행렬 
        # 각 문서는 n개의 단어 빈도를 통해 표현됨 
        # 이것을 k개의 차원으로 축소하는 것 

# 사이킷런 Truncated SVD클래스를 이용 
    # 문서의 LSA를 지원 
        # PCA 클래스에서 희소 행렬을 직접 연산하지 X
        # TruncateSVD는 바로 연산이 가능 --> toarray() 메서드 쓸 필요 없음 
        # TruncatedSVD는 transform() / fit_transform() 메서드 호출
            # fit_transform(): 학습 데이터셋에 대해 
            # transform() : 테스트 데이터에 대해 사용 
            # TruncatedSVD: 두 메서드에 대해 위와 같은 축소된 분석 행렬 반환 
                # 축소된 차원의 문서 행렬 반환 
                # 이때, 반환된 행렬은 위 절단된 SVD의   
                # UkSk : 학습 데이터에 대한 fit_transform 의 반환값
                # X' =XVk: 테스트 데이터에 대한 transform 의 반환값 

        # 축소된 차원 - 잠재된 의미 나타냄 
        # 문서와 단어를 연결
            # 1. 문서 벡터 m개의 차원
            # 2. 사용된 단어의 빈도에 기반해 문서를 표현
            # 3. k개의 차원으로 축소 
            # 4. k개의 잠재된 의미의 비중으로 각 문서를 표현
            # --> k개의 축소된 차원은 각각 잠재된 의미 표현 
            # 이를 단어에 대해서 적용하면, 각 단어도 k개의 잠재된 의미의 비중으로 표현돼 있음 
            # 단어들에 대해 잠재된 의미의 비중 
                # Sk x Sk^를 구하면 됨 
                # Sk - TruncatedSVD객체의 singularvalues 속성 
                # Sk^T - TruncatedSVD객체의 components_ 속성 
                # np.diag(svd.singularvalues).dot(svd.components_)  
```

### 6.3.4 단어 간 의미 유사도 분석
* LSA를 사용하면 -> 문서 간 유사도 & 단어 간 유사도 구할 수 있음 
* 카운트 벡터 행렬을 단어의 관점에서 보면 
    * 하나의 단어 --> 모든 문서에 나타난 빈도 
    * 문서에 나타나는 빈도 분포에 단어 간의 유사도 구할 수 있음 
    * LSA에서 문서가 축소된 의미의 비중으로 표현됨 
    * 단어 간의 의미적인 유사도 구할 수 있음 
    * 각, 단어의 의미 분포는 주어진 말뭉치에 한정됨 

* 토픽 
    * 토픽은 단어들의 비중으로 표현됨 
    * 전처리 과정을 단어 관점에서 다시 할 필요가 있음 

```python
#-----------------------------------------------------
# 잠재 토픽 분석 
    # TFIDF 벡터를 생성, 최대 단어 수 1,000개로 한정 
    # min_df, max_df 값 각각 5, 0.5로 지정 
    # TruncatedSVD 이용해 100개의 차원으로 축소 
    # 단어들의 잠재된 의미의 비중 - S x Vk^T 
        # np.diag(svd.singularvalues).dot(svd.components_)
        # svd.components_ shape : (100,1000) 
            # 변환된 단어 - 잠재의미 행렬에 대해 전치 행렬 - (1000,100)
            # 1,000개의 단어에 대해 100개의 내재된 의미를 이용해 유사도 구함 
#-----------------------------------------------------
tfidf = TfidfVectorizer(max_features=1000, min_df=5, max_df=0.5)
X_train_tfidf = tfidf.fit_transform(X_train) # train set을 변환

svd = TruncatedSVD(n_components=100, random_state=1) #압축할 component의 수 지정
X_train_lsa = svd.fit_transform(X_train_tfidf)

print('#components_의 shape:', svd.components_.shape)
print('#singular_values_의 shape:', svd.singular_values_.shape)
t_words = np.diag(svd.singular_values_).dot(svd.components_).T
print('#변환된 단어-잠재의미 행렬의 shape:', t_words.shape)

# space에 해당하는 벡터를 가져옴
source = t_words[np.where(tfidf.get_feature_names_out() == 'space')[0][0]]

sim_result = cosine_similarity([source], t_words) #변환된 count vector와 기존 값들과의 similarity 계산

print("#Top 20 유사도(tfidf):", sorted(sim_result[0].round(2), reverse=True)[:20])
sim_index = (-sim_result[0]).argsort()[:20]
print('#Top 20 유사 뉴스의 인덱스(tfidf):', sim_index)
sim_labels = [tfidf.get_feature_names_out()[i] for i in sim_index]
print('#Top 20 유사 뉴스의 카테고리(tfidf):', sim_labels)

# 결과 해석 
    # Top20 유사 뉴스의 카테고리(tfidf): ['space', 'exploration', 'sci', 'station', 'office', 'propulsion', 'reports', 'planetary', 'shuttle', 'national', 'astro', 'international', 'operations', 'missions', 'technical', 'satellites', 'probes', 'telescope', 'orbiter', 'nasa']
    # 'space'와 유사한 단어들을 검색, 자신에 대해 유사도 1이 나옴 
    # 그 외, 'exploration','sci','station' 등이 순서대로 출력 
    # 이 유사도는 주어진 문서 집합에서 함께 등장한 정도에 따라 계산됨, 우리가 생각하는 단어의 의미와 차이 있음 '
    # 말뭉치 내, 단어들이 어떻게 유사한 의미 갖는지 분석하고 싶을 때 이 방법을 사용 
```

## 6.3 tSNE를 이용한 시각화와 차원축소의 효과
* tSNE
    * 시각화를 위해 사용되는 비지도 학습 비지도 학습 알고리즘 
    * 다차원 데이터 사이의 거리를 가장 잘 보존하는 2차원 좌표를 찾기 위해 사용됨 
    * 즉 ,2차원 공간으로 데이터 간의 거리를 매핑 
    * --> 시각적으로 이를 쉽게 파악할 수 있도록 해줌 
    * tSNE를 이용해 그린 그래프에서 모양이 좋게 나타나다면 
    * --> 원래의 다차원 데이터 간의 거리도 의미 있게 잘 계산됨 

    * 의미를 이용해 문서들 간의 거리 계산 
    * 원래의 TFIDF를 이용한 문서 간 거리에 비해 문서의 유형을 더 잘 표현한다고 할 수 있을까?
    * 에제 데이터 구성 (20 뉴스그룹- 4개의 그룹 선택 )
        * 동일한 뉴스그룹 문서 -> 다른 문서들에 비해 거리가 가까울 것 
        * 즉, 잠재된 의미들이 문서의 유형 잘 파악한다면 
        * --> 2차원 공간에 문서 시각화했을 때, 같은 카테고리 문서, 서로 가깝게 분포 
        * 이를 확인하기 위해, 1) TFIDF -> t-SNE로 시각화 2) LSA -> t-SNE 시각화 비교

```python
#-----------------------------------------------------
# t-SNE
    # 차원을 축소하는 알고리즘 
    # 사이킷런 tSNE지원- TSNE 클래스 

    # TSNE 클래스 
        # TruncatedSVD와 유사 
        # 시각화는 위 토픽 분석에 사용한 데이터 그대로 활용
        # 객체를 선언한 후 2차원 좌표로 변환
        # 그 후 두 개의 축을 이용해 시각화 함 
        # 문서 유형이 그래프에 표현되도록 plt.text()를 이용해 문서의 카테고리(숫자)를 그래프에 출력 
#-----------------------------------------------------
from sklearn.manifold import TSNE

tfidf = TfidfVectorizer(tokenizer=tokenizer)
X_train_tfidf = tfidf.fit_transform(X_train) # train set을 변환
X_test_tfidf = tfidf.transform(X_test) # test set을 변환

tsne = TSNE(n_components=2, random_state=7)
tsne_tfidf = tsne.fit_transform(X_train_tfidf)
print('TSNE dimension:', tsne_tfidf.shape)

def tsne_graph(tsne_2, label, lim=None):
    colors = {0:'blue', 1:'red', 2:'green', 3:'purple'}
    x = tsne_2[:,0] #압축된 첫 차원을 x축으로 이용
    y = tsne_2[:,1] #압축된 둘째 차원은 y축으로 이용
    plt.figure(figsize=(15,10))
    if lim == None:
        lim = [min(x), max(x), min(y), max(y)]
    plt.xlim(lim[0], lim[1])
    plt.ylim(lim[2], lim[3])
    #for i in range(500):
    for i in range(len(x)):
        #각 값에 대해 y값 즉 label에 따라 색을 바꿔가며 text로 그래프에 출력
        if (lim[0] < x[i] < lim[1]) and (lim[2] < y[i] < lim[3]):
            plt.text(x[i], y[i], label[i], color = colors[label[i]])
    plt.show()
    
tsne_graph(tsne_tfidf, y_train, (-4, 5, -5, 5))

# 결과 해석
    # 각 카테고리에 해당하는 문서들이 유형에 고나계없이 특이하게 분포 
    # 패턴이 있지만, 그래프를 통해, 카테고리 구분하기 어려움 
    # --> 따라서, LSA를 이용해 차원 축소하고 --> t-SNE로 다시 시각화 
```
